```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import warnings

# This hides annoying warning messages so your screen stays clean
warnings.filterwarnings('ignore') 
print("Libraries loaded successfully!")
```

```python
# The CSV file is in the archive folder, one level up from IDS_Project
df = pd.read_csv(r'C:\Users\darsh\.gemini\antigravity\scratch\archive\NF-UQ-NIDS-v2.csv', nrows=100000)

print(f"Loaded {df.shape[0]} rows and {df.shape[1]} columns.")
# Look at the first 5 rows to see what the raw data looks like
df.head()
```

```python
# 1. Drop any rows that have missing data
df = df.dropna()

# 2. In some datasets, attack labels are strings (e.g., 'Benign', 'BruteForce'). 
# We need to ensure we have a binary 'Label' column (0 for Normal, 1 for Attack).
# If your dataset has an 'Attack' column with 1s and 0s, we use that.
if 'Attack' in df.columns:
        df['Label'] = df['Attack']

print("Data cleaned. Current shape:", df.shape)
```

```python
# --- Feature 1: IP Diversity (Distributed Attack Detection) ---
# How many unique Source IPs are targeting a specific Destination IP?
ip_diversity = df.groupby('IPV4_DST_ADDR')['IPV4_SRC_ADDR'].nunique().reset_index()
ip_diversity.columns = ['IPV4_DST_ADDR', 'Target_IP_Diversity']

# Merge this new behavior back into our main dataset
df = pd.merge(df, ip_diversity, on='IPV4_DST_ADDR', how='left')

# --- Feature 2: Attack Frequency (Aggressive Attack Detection) ---
# How many total connections has this specific Source IP made?
ip_frequency = df.groupby('IPV4_SRC_ADDR').size().reset_index(name='Source_Connection_Count')
df = pd.merge(df, ip_frequency, on='IPV4_SRC_ADDR', how='left')

print("Behavioral features successfully engineered!")
```

```python
# Separate the "Answers" (Labels) from the "Questions" (Features)
y = df['Label'] # This is what we want to predict (0 or 1)

# Drop columns that are strings/IPs (ML models only read math/numbers)
# We also drop the Label so the model doesn't cheat by looking at the answer
columns_to_drop = ['IPV4_SRC_ADDR', 'IPV4_DST_ADDR', 'Attack', 'Label', 'Dataset']
X = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to a DataFrame so it's easy to read
X_final = pd.DataFrame(X_scaled, columns=X.columns)

print("Data successfully normalized and ready for model training!")
```

```python
from sklearn.model_selection import train_test_split

# Split the data: 80% for training, 20% for testing
# random_state=42 ensures you get the exact same split every time you run this
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
import time

print("Training Random Forest... This might take a minute.")
start_time = time.time()

# Initialize the model with 100 "decision trees"
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

# Fit (Train) the model on the data
rf_model.fit(X_train, y_train)

end_time = time.time()
print(f"Training Complete! It took {round(end_time - start_time, 2)} seconds.")
```

```python
from sklearn.ensemble import RandomForestClassifier
import time
import numpy as np

# Fix: replace infinite values and clip extreme values before training
X_train_clean = X_train.replace([np.inf, -np.inf], np.nan).fillna(0)
X_test_clean = X_test.replace([np.inf, -np.inf], np.nan).fillna(0)

# Also clip any values that are too large for float32
X_train_clean = X_train_clean.clip(lower=-1e10, upper=1e10)
X_test_clean = X_test_clean.clip(lower=-1e10, upper=1e10)

print("Training Random Forest... This might take a minute.")
start_time = time.time()

rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_model.fit(X_train_clean, y_train)

end_time = time.time()
print(f"Training Complete! It took {round(end_time - start_time, 2)} seconds.")
```

```python

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns

# Ask the trained model to predict the answers for the test data
y_pred = rf_model.predict(X_test_clean)

# 1. Calculate Overall Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"--- Model Accuracy: {accuracy * 100:.2f}% ---")

# 2. Detailed Classification Report (Recall, Precision)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# 3. Generate a Visual Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted Label (0=Normal, 1=Attack)')
plt.ylabel('True Label (0=Normal, 1=Attack)')
plt.title('Random Forest Confusion Matrix')
plt.show()
```

```python
import numpy as np
X_train_dl = np.expand_dims(X_train_clean, axis=2)
X_test_dl = np.expand_dims(X_test_clean, axis=2)
print(f"X_train_dl shape: {X_train_dl.shape}")
print(f"X_train_dl dtype: {X_train_dl.dtype}")
print(f"y_train dtype: {y_train.dtype}")
print(f"y_train unique values: {np.unique(y_train)}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout

# Convert labels to binary (0 = Benign, 1 = Attack)
# Assuming 'Benign' is the label for normal traffic
y_train_bin = (y_train != 'Benign').astype(int)
y_test_bin = (y_test != 'Benign').astype(int)

print("Building CNN Model...")

# 1. Define the CNN Architecture
cnn_model = Sequential([
        # The Convolutional layer "scans" the data features
        Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train_dl.shape[1], 1)),
        MaxPooling1D(pool_size=2), # Compresses the data to highlight the most important parts
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.3), # This prevents "overfitting" by randomly turning off neurons during training
        Dense(1, activation='sigmoid') # Final output: 0 (Normal) or 1 (Attack)
])

# 2. Compile the Model
cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("Training CNN... Grab a coffee, this will take a moment.")
# Train for 5 'epochs' (5 complete passes through the dataset)
cnn_history = cnn_model.fit(X_train_dl, y_train_bin, epochs=5, batch_size=64, validation_split=0.2)
```

```python
from tensorflow.keras.layers import LSTM

print("Building LSTM Model...")

# 1. Define the LSTM Architecture
lstm_model = Sequential([
        LSTM(64, input_shape=(X_train_dl.shape[1], 1), return_sequences=False),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
])

# 2. Compile the Model
lstm_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("Training LSTM... This usually takes slightly longer than the CNN.")
# NOTE: Using y_train_bin because y_train contains string labels and Keras needs 0/1 integers
lstm_history = lstm_model.fit(X_train_dl, y_train_bin, epochs=5, batch_size=64, validation_split=0.2)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

print("Generating detailed LSTM Evaluation...")

# The LSTM outputs a probability between 0 and 1 (e.g., 0.85 means 85% sure it's an attack).
# We convert any probability > 0.5 into a solid '1' (Attack), and anything lower to '0' (Normal).
lstm_probabilities = lstm_model.predict(X_test_dl)
y_pred_lstm = (lstm_probabilities > 0.5).astype(int)

# 1. Detailed Classification Report
print("\nLSTM Classification Report:")
# IMPORTANT: y_test_bin and y_train_bin were likely created by you previously to handle the string labels.
# Use y_test_bin if you created it, otherwise use y_test. Let's assume you need to use y_test_bin, but try y_test if it exists as 0/1. Let's just use y_test_bin:
print(classification_report(y_test_bin, y_pred_lstm))

# 2. Visual Confusion Matrix
cm_lstm = confusion_matrix(y_test_bin, y_pred_lstm)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_lstm, annot=True, fmt='d', cmap='Reds', cbar=False)
plt.xlabel('Predicted Label (0=Normal, 1=Attack)')
plt.ylabel('True Label (0=Normal, 1=Attack)')
plt.title('LSTM Confusion Matrix - Brute Force Detection')
plt.show()
```

```python

from tensorflow.keras.callbacks import EarlyStopping

print("Retraining LSTM with Early Stopping optimization...")

# EarlyStopping monitors the 'validation loss'. If it doesn't improve for 3 cycles (patience=3), 
# it restores the best version of the model.
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# We use the same architecture, but train it with our new safety net
optimized_history = lstm_model.fit(
        X_train_dl, 
        y_train_bin, # IMPORTANT: Using y_train_bin because y_train has string labels
        epochs=20, # We allow it to run longer now, knowing it will stop itself if needed
        batch_size=64, 
        validation_split=0.2,
        callbacks=[early_stop]
)

print("\nOptimized Training Complete!")
```
